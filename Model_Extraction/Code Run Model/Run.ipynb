```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Load the features and labels from .npy files
features = np.load('/content/saved_features_eva02_large_patch14_448.mim_m38m.npy')
labels = np.load('/content/labels_eva02_large_patch14_448.mim_m38m.npy')
```

```python

# Define a function to retrieve the most similar images with their indices
def retrieve_image(query_feature, features, labels, top_k=5):
    # Compute cosine similarity between the query and all other features
    similarities = cosine_similarity(query_feature.reshape(1, -1), features).flatten()

    # Get the indices of the top_k most similar images
    top_indices = np.argsort(similarities)[::-1][:top_k]

    # Retrieve the corresponding indices, labels, and similarity scores for the top_k results
    results = [(index, labels[index], similarities[index]) for index in top_indices]

    return results

# Example usage
query_index =1200  # Select an index from the dataset to use as a query
query_feature = features[query_index]

# Retrieve the most similar images
similar_images = retrieve_image(query_feature, features, labels, top_k=5)

# Print the results
for index, label, similarity in similar_images:
    print(f"Index: {index}, Label: {label}, Similarity: {similarity:.4f}")
```
